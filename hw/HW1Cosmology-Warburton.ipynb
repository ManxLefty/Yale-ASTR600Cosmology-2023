## Problem 1

Consider the integral 
$$
\int_0^z \frac{dz}{\left[\Omega (1+z)^3 + (1-\Omega) (1+z)^{3/2}\right]^{1/2}}
$$
where $\Omega$ is a parameter between 0 and 1. Note that this integral defines a function 
of $z$.

Write code to numerically integrate this integral for a given value of $\Omega$ and $z$.
Plot the result as a function of $z$ (between $0<z<1$) for $\Omega=0, 0.3, 0.7, 1$.
Make sure to label the axes of your plot.

Also include on your plot the analytic results for $\Omega=0$ and $\Omega=1$.

You can use any approach to integrate the function that you like (i.e. you do not 
need to write the integration routine from scratch). 

## Solution

```python
# Our lovely imports

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

### Numerical integration

```python
# Define function to build integrand
def integrand(z, Omega):
    return 1 / np.sqrt(Omega * (1 + z)**3 + (1 - Omega) * (1 + z)**(3/2))

# Range of z values (0<z<1) over which to integrate
z_vals = np.linspace(0, 1, 100)

# Values of Omega
Omega_vals = [0, 0.3, 0.7, 1]

# Numerical integration for each value of Omega
results = {}  # Dictionary to collect results
for Omega in Omega_vals:
    results[Omega] = [quad(integrand, 0, z, args=(Omega,))[0] for z in z_vals]
```

### Analytical integration solutions

When $\Omega$ = 0, the integral becomes: 
$$\int_0^z \frac{1}{\left[(1+z)^{3/2}\right]^{1/2}} dz = \int_0^z \frac{1}{(1+z)^{3/4}} dz = \int_0^z (1+z)^{-3/4} dz = 4(1+z)^{1/4} - 4$$

When $\Omega$ = 1, the integral becomes:
$$\int_0^z \frac{1}{\left[(1+z)^3\right]^{1/2}} dz = \int_0^z \frac{1}{(1+z)^{3/2}} dz = \int_0^z (1+z)^{-3/2} dz = -2(1+z)^{-1/2} + 2$$

```python
# Analytical integration results for Omega=0,1
analytic_0 = [4*(1 + z)**(1/4) - 4 for z in z_vals]
analytic_1 = [-2*(1 + z)**(-1/2) + 2 for z in z_vals]
```

### Plot the results

```python
# Plot the results

# Set figure
fig, ax = plt.subplots(figsize=(8,6))

# Plot numerical integral result for each value of Omega
for Omega in Omega_vals:
    plt.plot(z_vals, results[Omega], label='$\Omega$ = ' + str(Omega), lw=2)
    
# Plot analytical integral results for Omega=0,1
plt.plot(z_vals, analytic_0, label='$\Omega$ = 0 (analytic)', ls=':', color='lightblue', lw=2)
plt.plot(z_vals, analytic_1, label='$\Omega$ = 1 (analytic)', ls=':', color='pink', lw=2)

# Label axes   
plt.xlabel('z', fontsize=16)
plt.ylabel('Integration Result', fontsize=16)

# Increse axis numbering text size
ax.tick_params(axis='both', which='major', labelsize=10)
ax.tick_params(axis='both', which='minor', labelsize=10)

# Plot features
plt.legend(fontsize=14)
plt.grid(True, color='lightgrey', ls='-.')
plt.title('Numerical Integration Exercise', fontsize=18)

# Save and show
plt.savefig("HW1IntegrationPlot.pdf", format="pdf", bbox_inches="tight", overwrite=True)
plt.show()
```
